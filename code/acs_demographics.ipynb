```python
#%pip install census

from census import Census
import pandas as pd

censuskey = "8105419cada33ca0aaa48b111b8c44b9484e286a"
```

```python

def demographics(variables_list, year = 2016):
    
    list_of_dfs = [pd.DataFrame(c.acs5.state_county_tract(variables_list,
                                                          state_fips=fip,
                                                          year = year,
                                                          county_fips=Census.ALL,
                                                          tract=Census.ALL))
                   for fip in fips]
    
    full_df = pd.concat(list_of_dfs)
                   
    return full_df
```

# Get State FIP Codes

```python

fips_url = 'https://gist.githubusercontent.com/dantonnoriega/bf1acd2290e15b91e6710b6fd3be0a53/raw/11d15233327c8080c9646c7e1f23052659db251d/us-state-ansi-fips.csv'

state_fips = pd.read_csv(fips_url)
state_fips = state_fips.rename(columns = {'stname': 'name', ' st': 'fip', ' stusps': 'abbr'})

fips = state_fips['fip'].astype(str)
```

```python
for i in range(0, len(fips)):
    if len(fips[i]) < 2:
        fips[i] = '0' + fips[i]
```

# Get A Few Demographic Variables For All States

```python
var_names = ["B03001_003E", "B01003_001E", "NAME", "B06012_002E", "B07007_005E"]

c = Census(censuskey)

all_dem = demographics(var_names)
```

```python

all_dem["perc_hispanic"] = all_dem["B03001_003E"]/all_dem["B01003_001E"]
all_dem["perc_bpl"] = all_dem["B06012_002E"]/all_dem["B01003_001E"]
all_dem["perc_mobile"] = all_dem["B07007_005E"]/all_dem["B01003_001E"]

all_dem["GEOID_pasted"] = all_dem.state.astype(str) + \
                        all_dem.county.astype(str) + all_dem.tract.astype(str)
```
